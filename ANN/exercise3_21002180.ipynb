```python
from __future__ import division, print_function, unicode_literals

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
import zipfile
import os

# Path to the ZIP file
zip_file_path = "E:\\ML\\ANN\\face_data.zip"

# Extracted directory
extracted_file = "E:\\ML\\ANN\\face_data"

os.makedirs(extracted_file, exist_ok=True)

# Extract the contents of the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_file)
```

```python
# path to the database - change it if needed
path = "E:\\ML\\ANN\\face_data\\face_data\\"

ids = range(1, 16) # 15 people
states = ['centerlight', 'glasses', 'happy', 'leftlight', 
          'noglasses', 'normal', 'rightlight','sad', 
          'sleepy', 'surprised', 'wink']

prefix = 'subject'
surfix = '.png' #file extension is png

# open one picture to get the image's size
fn = prefix + '01.' + states[0] + surfix
im = cv2.imread(path + fn, 0)

h = im.shape[0] # hight 
w = im.shape[1] # width

D = h * w
N = len(states)*15 
print(N, D, h, w)

X = np.zeros((D, N))

# collect all data
cnt = 0 

# there are 15 people
for person_id in range(1, 16):
    for state in states:
        
        # get name of each image file
        fn = path + prefix + str(person_id).zfill(2) + '.' + state + surfix
        
        # open the file and read as grey image
        tmp = cv2.imread(fn, cv2.IMREAD_GRAYSCALE)
        
        # then add image to dataset X
        X[:, cnt] = tmp.reshape(D)
        cnt += 1 
```

#  Giảm chiều dữ liệu xuống 125

```python
# Doing PCA, note that each row is a datapoint
from sklearn.decomposition import PCA

# remain dim. k = 125 - change it! 
pca = PCA(n_components=125) 

# then apply to data X
X_pca = pca.fit_transform(X.T)

# then build projection matrix 
U = pca.components_.T
```

```python
# then reshape new dataset (reduced dim.) to be new images and save
# path to save reduced dim. images 
path_save = "E:\\ML\\ANN\\face_data\\data\\results\\"
for i in range(U.shape[1]):
    plt.axis('off')
    f1 = plt.imshow(U[:, i].reshape(h, w), interpolation='nearest')
    f1.axes.get_xaxis().set_visible(False)
    f1.axes.get_yaxis().set_visible(False)

    plt.gray()
    fn = path_save + 'eigenface' + str(i).zfill(2) + '.png'
    plt.savefig(fn, bbox_inches='tight', pad_inches=0)
    

# test results by showing some images
# See reconstruction of first 6 persons 
for person_id in range(1, 7):
    for state in ['centerlight']:
        fn = path + prefix + str(person_id).zfill(2) + '.' + state + surfix
        im = cv2.imread(fn,0)
        plt.axis('off')

        f1 = plt.imshow(im, interpolation='nearest')
        f1.axes.get_xaxis().set_visible(False)
        f1.axes.get_yaxis().set_visible(False)
        plt.gray()
        fn = 'ori' + str(person_id).zfill(2) + '.png'
        plt.savefig(fn, bbox_inches='tight', pad_inches=0)
        plt.show()
        
        # reshape and subtract mean, don't forget 
        x = im.reshape(D, 1) - pca.mean_.reshape(D, 1)
        
        # encode
        z = U.T.dot(x)
        
        #decode
        x_tilde = U.dot(z) + pca.mean_.reshape(D, 1)

        # reshape to orginal dim
        im_tilde = x_tilde.reshape(h, w)
        plt.axis('off')

        f1 = plt.imshow(im_tilde, interpolation='nearest')
        f1.axes.get_xaxis().set_visible(False)
        f1.axes.get_yaxis().set_visible(False)
        plt.gray()
#        fn = 'res' + str(person_id).zfill(2) + '.png'
#        plt.savefig(fn, bbox_inches='tight', pad_inches=0)
        plt.show()
```

# Áp dụng các phương pháp phân loại nhiều lớp: Multinomial Logistic Regression, Naïve Bayes phù hợp và ANN (đã có code) để phân loại, tỷ lệ train:test là 0.7:0.3.

```python
import os
import numpy as np

# Tạo một list để lưu trữ các nhãn
labels = []

for file_name in os.listdir(path):
    if file_name.endswith('.png'):
        # Lấy tên của nhãn từ tên tệp ảnh
        label = file_name.split('.')[1]  
        labels.append(label)

# Chuyển đổi list nhãn thành numpy array
y_labels = np.array(labels)

print(y_labels)
```

## Sử dụng Multinomial Logistic

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y_labels, test_size=0.3, random_state=42)
```

```python
# Huấn luyện mô hình Logistic Regression
log_reg_model = LogisticRegression(multi_class='multinomial',solver='sag',  max_iter=1000)
log_reg_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Dự đoán và đánh giá hiệu suất của mô hình Logistic Regression
log_reg_pred = log_reg_model.predict(X_test)
print(log_reg_pred)
```

```python
log_reg_accuracy = accuracy_score(y_test, log_reg_pred)
print("Logistic Regression Accuracy:", log_reg_accuracy)
```

## Sử dụng Naïve Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_bayes_model = GaussianNB()
naive_bayes_model.fit(X_train, y_train)
```

```python
# Dự đoán và đánh giá hiệu suất của mô hình Naive Bayes
naive_bayes_pred = naive_bayes_model.predict(X_test)
print(naive_bayes_pred)
```

```python
naive_bayes_accuracy = accuracy_score(y_test, naive_bayes_pred)
print("Naive Bayes Accuracy:", naive_bayes_accuracy)
```

## Sử dụng ANN

```python
from sklearn.neural_network import MLPClassifier

ann = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
ann.fit(X_train, y_train)
```

```python
y_pred_ann = ann.predict(X_test)
print(y_pred_ann)
```

```python
accuracy_ann = accuracy_score(y_test, y_pred_ann)
print("ANN Accuracy:", accuracy_ann)
```
