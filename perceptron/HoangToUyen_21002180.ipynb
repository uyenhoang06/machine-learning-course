```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
df = pd.read_csv("E:\\ML\\perceptron\\framingham.csv")
df.head()
```

# 1. Hãy xử lý sơ bộ dữ liệu như trong yêu cầu của phần hồi quy logistic. Chia tập dữ liệu thành Traning:Validation = 7:3.

```python
df.isnull().sum()
```

```python
print("Số lượng bản ghi trong DataFrame:", len(df))
print("Kích thước DataFrame:", df.shape)
```

```python
# loại bỏ các thành phần N/A
df = df.dropna(how = "any", axis = 0)
```

```python
df.isnull().sum()
```

```python
print("Số lượng bản ghi trong DataFrame lúc này:", len(df))
print("Kích thước DataFrame:", df.shape)
```

```python
# chia dữ liệu thành tập training và validation
X = df.drop(columns=['TenYearCHD'])
y = df['TenYearCHD']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_val.shape)
print(y_val.shape)
```

# 2. Thực hiện phân loại bằng phương pháp Perceptron. Tính các độ đo Accuracy, Precision và Recall để đánh giá kết quả.

## Sử dụng thư viện

```python
from sklearn.linear_model import Perceptron 
```

```python
y_val_per = y_val.replace(0, -1)
y_train_per = y_train.replace(0, -1)
```

```python
perceptron = Perceptron(fit_intercept=True, random_state = 2) 
perceptron.fit(X_train, y_train_per)
```

```python
print(perceptron.coef_)
```

```python
print(perceptron.intercept_)
```

```python
# Dự đoán trên tập Validation
y_pred = perceptron.predict(X_val)
print(y_pred)
```

```python
# Tính độ chính xác, recall và precision đối với phương pháp Perceptron 
accuracy = accuracy_score(y_val_per, y_pred)
recall = recall_score(y_val_per, y_pred)
precision = precision_score(y_val_per, y_pred)
f1 = f1_score(y_val_per, y_pred)

print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1-Score:", f1)
```

# 3. Thực hiện phân loại bằng phương pháp Hồi quy Logistic trên cùng bộ dữ liệu training:validation đã có ở ý trên. Tính các độ đo Accuracy, Precision và Recall và so sánh kết quả.

```python
# Huấn luyện mô hình Logistic Regression trên tập Training
logReg = linear_model.LogisticRegression(penalty=None, max_iter=10000, fit_intercept=False, random_state = 2)
logReg.fit(X_train, y_train)
```

```python
# In ra các hệ số của mô hình
print("Các hệ số của mô hình:", logReg.coef_)

# In ra hệ số tự do của mô hình
print("Hệ số tự do của mô hình:", logReg.intercept_)
```

```python
# Dự đoán trên tập Validation
y_prob = logReg.predict_proba(X_val)[:, 1]
y_pred = logReg.predict(X_val)

y_result_df = pd.DataFrame({'y_prob': y_prob, 'y_pred': y_pred})
print(y_result_df.head(10))
```

```python
# Tính độ chính xác, recall và precision đối với phương pháp hòi quy Logistic 
accuracy = accuracy_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)

print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1-Score:", f1)
```

# 4. Hãy giải thích về kết quả thu được của các mô hình cũng như nhận xét trên độ chính xác của mỗi mô hình.

Random state: là một giá trị khởi tạo ngẫu nhiên cho các thuật toán học máy. Việc sử dụng random state giúp đảm bảo tính khách quan và minh bạch cho quá trình học tập, đồng thời giúp kiểm soát độ biến động của mô hình.

- Khi sử dụng một số giá trị random_state giống nhau, thì độ chính xác cũng như các chỉ số recall, precision, f1-score của 2 mô hình là gần gống nhau.
- Đối với mô hình Logistic, khi khi thay đổi các giá trị random_state thì các chỉ số về accuracy, recall, precision, f1-score là gần giống nhau, không có quá nhiều sự chênh lệch.
- Còn đối với Perceptron, khi thay đổi các giá trị random_state đôi khi các chỉ số thu được lại chênh nhau rất lớn.

Giải thích:
- Mô hình Logistic có tính ổn định cao hơn. Trong Logistic Regression, mục tiêu là tìm ra một hàm sigmoid (logistic) sao cho dự đoán của mô hình càng gần nhãn thực tế của dữ liệu càng tốt. Thuật toán sử dụng một hàm mất mát (ví dụ: Cross-Entropy Loss) để đo lường sự khác biệt giữa dự đoán của mô hình và các nhãn thực tế của dữ liệu. Quá trình tối ưu hóa cố gắng điều chỉnh các trọng số của mô hình sao cho hàm mất mát là nhỏ nhất có thể, thường thông qua các phương pháp tối ưu hóa như Gradient Descent. Trong quá trình này, các trọng số của mô hình được điều chỉnh liên tục dựa trên đạo hàm của hàm mất mát, và quá trình này tiếp tục cho đến khi đạt được một điểm cực tiểu hoặc gần cực tiểu của hàm mất mát. Do quá trình này là một quá trình liên tục và không phụ thuộc nhiều vào thứ tự của dữ liệu, nó thường cho ra kết quả ổn định hơn khi sử dụng các giá trị random_state khác nhau.

- Còn mô hình Perceptron là một thuật toán học tuyến tính. Perceptron phân loại dữ liệu dựa trên việc tìm những đường thẳng, siêu phẳng để phân chia. Các Trọng số trong mô hình được cập nhật dựa trên các điểm dữ liệu một cách tuần tự và không sử dụng lại dữ liệu huấn luyện sau nhiều lần như trong Logistic Regression, quá trình này tiếp tục cho đến khi không còn sai số nào hoặc số lần lặp vượt qua một ngưỡng cụ thể. Chính vì vậy sự ổn định của Perceptron có thể bị ảnh hưởng bởi việc khởi tạo thứ tự các điểm dữ liệu để đưa vào huân luyện. Một thứ tự sắp xếp bộ dữ liệu tốt có thể dẫn đến việc hội tụ nhanh chóng đến một giải pháp tốt, trong khi một bộ thứ tự khởi tạo không tốt có thể dẫn đến hội tụ đến một giải pháp kém hoặc không hội tụ đến một kết quả tối ưu.
