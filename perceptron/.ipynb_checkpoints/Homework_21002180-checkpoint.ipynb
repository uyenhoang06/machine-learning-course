# 1. Exercise 1

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)
means = [[2, 2], [4, 2]]
cov = [[.3, .2], [.2, .3]]

N = 30
X0 = np.random.multivariate_normal(means[0], cov, N).T
X1 = np.random.multivariate_normal(means[1], cov, N).T

print(X0.shape)
print(X0)
```

```python
X = np.concatenate((X0, X1), axis = 1)
y = np.concatenate((np.ones((1, N)), -1*np.ones((1, N))), axis = 1)

print(X)
print(X.shape)
```

```python
# Xbar
X = np.concatenate((np.ones((1, 2*N)), X), axis = 0)
print(X.shape)
```

```python
print(X)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ani = plt.cla()

#plot points
ani = plt.plot(X0[0, :], X0[1, :], 'b^', markersize = 8, alpha = .8)
ani = plt.plot(X1[0, :], X1[1, :], 'ro', markersize = 8, alpha = .8)
ani = plt.axis([0 , 6, -2, 4])
plt.show()
```

```python
# Define h_w(x):= W^T.x + w_0 = \bar{W}^T . \bar{x}
def h(w, x):
    return np.sign(np.dot(w.T, x))
```

```python
#Stop condition
def has_converged(X, y, w):
    return np.array_equal(h(w, X), y) #True if h(w, X) == y else False
```

```python
def perceptron(X, y, w_init):
    w = [w_init]
    N = X.shape[1]
    mis_points = [] # set of miss position points
    
    while True:
        # mix data
        mix_id = np.random.permutation(N)
        
        for i in range(N):
            xi = X[:, mix_id[i]].reshape(3, 1)
            yi = y[0, mix_id[i]]
            
            if h(w[-1], xi)[0] != yi:
                mis_points.append(mix_id[i])
                w_new = w[-1] + yi*xi
                
                w.append(w_new)
        
        if has_converged(X, y, w[-1]):
            break
    return (w, mis_points)
```

```python
d = X.shape[0]
w_init = np.random.randn(d, 1)
(w, m) = perceptron(X, y, w_init)
print(w[-1])
```

```python
print(d)
```

```python
print(y)
print(y.shape)
```

```python
def draw_line(w):
    w0, w1, w2 = w[0], w[1], w[2]
    if w2 != 0:
        x11, x12 = -100, 100
        return plt.plot([x11, x12], [-(w1*x11 + w0)/w2, -(w1*x12 + w0)/w2], 'k')
    else:
        x10 = -w0/w1
        return plt.plot([x10, x10], [-100, 100], 'k')
```

```python
## Visualization
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation, PillowWriter
import PIL

def viz_alg_1d_2(w):
    it = len(w)
    fig, ax = plt.subplots(figsize=(5, 5))
    
    def update(i):
        ani = plt.cla()
        #points
        ani = plt.plot(X0[0, :], X0[1, :], 'b^', markersize = 8, alpha = .8)
        ani = plt.plot(X1[0, :], X1[1, :], 'ro', markersize = 8, alpha = .8)
        ani = plt.axis([0 , 6, -2, 4])
        i2 = i if i < it else it-1
        ani = draw_line(w[i2])
        
        if i < it-1:
        # draw one misclassified point
            circle = plt.Circle((X[1, m[i]], X[2, m[i]]), 0.15, color='k', fill = False)
            ax.add_artist(circle)
        
        cur_axes = plt.gca()
        cur_axes.axes.get_xaxis().set_ticks([])
        cur_axes.axes.get_yaxis().set_ticks([])
        label = 'PLA: iter %d/%d' %(i2, it-1)
        ax.set_xlabel(label)
        return ani, ax
                
    anim = FuncAnimation(fig, update, frames=np.arange(0, it + 2), interval=1000)
    # save
    anim.save('E:\\ML\\perceptron\\pla_vis.gif', dpi = 100, writer = 'imagemagick')
    plt.show()

viz_alg_1d_2(w)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot
import matplotlib.pyplot
```

```python
main_df = pd.read_csv("E:\\ML\\perceptron\\sonar.all-data.csv", header=None)
main_df
```

```python
main_df[60].value_counts().plot(kind='barh')
```

```python
y_df = main_df[60]
targes_label = {'M': 1,'R': -1}
targes_df = [targes_label[item] for item in y_df]
print(targes_df)
```

```python
y = np.array(targes_df)
print(y.shape)
```

```python
inputs_df = main_df.drop(60, axis=1)
x0 = np.ones((inputs_df.shape[0], 1))

print(x0.shape)
X = np.concatenate((x0, inputs_df), axis = 1)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=42)
```

```python
y_train = y_train.reshape(1, len(y_train))
print(y_train.shape)
y_test = y_test.reshape(1, len(y_test))
print(y_test.shape)
```

```python
X_train  = X_train.T
print(X_train)
print(X_train.shape)
```

```python
X_test  = X_test.T
print(X_test)
print(X_test.shape)
```

```python
# Define h_w(x):= W^T.x + w_0 = \bar{W}^T . \bar{x}
def h(w, x):
    return np.sign(np.dot(w.T, x))

#Stop condition
def has_converged(X, y, w):
    return np.array_equal(h(w, X), y) #True if h(w, X) == y else False

def perceptron(X, y, w_init):
    w = [w_init]
    N = X.shape[1]
    mis_points = [] # set of miss position points
    
    while True:
        # mix data
        mix_id = np.random.permutation(N)
        
        for i in range(N):
            xi = X[:, mix_id[i]].reshape(61, 1)
            yi = y[0, mix_id[i]]
            
            if h(w[-1], xi)[0] != yi:
                mis_points.append(mix_id[i])
                w_new = w[-1] + yi*xi
                
                w.append(w_new)
        
        if has_converged(X, y, w[-1]):
            break
    return (w, mis_points)


d = X_train.shape[0]
w_init = np.random.randn(d, 1)
(w, m) = perceptron(X_train, y_train, w_init)
print(w[-1])
```

```python
y_pred = h(w[-1], X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

y_test = y_test.reshape(y_test.shape[1], 1)
y_pred = y_pred.reshape(y_pred.shape[1], 1)


# Tính độ chính xác, recall và precision đối với phương pháp hòi quy Logistic 
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1-Score:", f1)
```

## Bộ dữ liệu ở ví dụ 3

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

cancer_data = datasets.load_breast_cancer()

# show to test record 5th
print(cancer_data.data[5])
print(cancer_data.data.shape)

targets = cancer_data.target

#target set
print(targets)
print(targets.shape)
```

```python
print(X_train)
```

```python
print(y_train)
```

### Giảm số chiều về 2 chiều

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Giảm chiều dữ liệu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(cancer_data.data)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(X_pca.shape)
```

```python
print(X_pca[1])
```

```python
transformed = pd.DataFrame(X_pca)
```

```python
plt.axis("off")

plt.scatter(transformed[targets==0][0], transformed[targets==0][1], s=9, label='0',
c='red', marker = 's')
plt.scatter(transformed[targets==1][0], transformed[targets==1][1], s=9, label='1',
c='green', marker="s")

plt.legend()
plt.show()
```

### Thực hiện Perceptron trên dữ liệu chưa giảm chiều

```python
from sklearn.linear_model import Perceptron 
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target,test_size=0.3,random_state=109)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
perceptron = Perceptron(fit_intercept=True, random_state = 2) 
perceptron.fit(X_train, y_train)
```

```python
print(perceptron.coef_)
```

```python
print(perceptron.intercept_)
```

```python
# Dự đoán trên tập Validation
y_pred = perceptron.predict(X_test)
print(y_pred)
```

```python
# Tính độ chính xác, recall và precision đối với phương pháp Perceptron 
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1-Score:", f1)
```

### Thực hiện hiện hồi quy Logistic trên dữ liệu chưa giảm chiều

```python
# Huấn luyện mô hình Logistic Regression trên tập Training
logReg = linear_model.LogisticRegression(penalty=None, max_iter=10000, fit_intercept=False, random_state = 2)
logReg.fit(X_train, y_train)
```

```python
# In ra các hệ số của mô hình
print("Các hệ số của mô hình:", logReg.coef_)

# In ra hệ số tự do của mô hình
print("Hệ số tự do của mô hình:", logReg.intercept_)
```

```python
# Dự đoán trên tập Validation
y_prob = logReg.predict_proba(X_test)[:, 1]
y_pred = logReg.predict(X_test)

y_result_df = pd.DataFrame({'y_prob': y_prob, 'y_pred': y_pred})
print(y_result_df.head(10))
```

```python
# Tính độ chính xác, recall và precision đối với phương pháp hòi quy Logistic 
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1-Score:", f1)
```
