### Sử dụng thư viện - cách 2

```python
# Áp dụng PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Tính tổng phần trăm phương sai giữ lại của từng thành phần chính
explained_variance_ratio_cumsum = np.cumsum(pca.explained_variance_ratio_)

# Chọn số lượng thành phần chính để giữ lại ít nhất 80% thông tin
n_components = np.argmax(explained_variance_ratio_cumsum >= 0.8) + 1

# Giảm chiều dữ liệu
pca = PCA(n_components=n_components)
X_pca_reduced = pca.fit_transform(X_scaled)

print("Số lượng thành phần chính được giữ lại:", n_components)
print("Tổng phần trăm phương sai giữ lại:", explained_variance_ratio_cumsum[n_components-1])
```

### Không sử dụng thư viện

```python
# Tính ma trận hiệp phương sai
def compute_covariance_matrix(data):
    covariance_matrix = np.cov(data, rowvar=False)
    return covariance_matrix

# Tính các vectơ riêng và giá trị riêng
def find_eigenvalues_and_eigenvectors(covariance_matrix):
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
    return eigenvalues, eigenvectors

# Sắp xếp các giá trị riêng theo thứ tự giảm dần và chọn các vectơ riêng tương ứng
def sort_eigenvalues_and_eigenvectors(eigenvalues, eigenvectors):
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]
    return sorted_eigenvalues, sorted_eigenvectors

# Chiếu dữ liệu vào các thành phần chính đã chọn
def project_data(data, eigenvectors, num_components):
    projection_matrix = eigenvectors[:, :num_components]
    projected_data = np.dot(data, projection_matrix)
    return projected_data

# Tính tổng phần trăm phương sai giữ lại của các thành phần chính
def compute_variance_retained(eigenvalues, num_components):
    total_variance = np.sum(eigenvalues)
    variance_retained = np.sum(eigenvalues[:num_components]) / total_variance
    return variance_retained
```

```python
# Tiêu chuẩn hóa dữ liệu
data_mean = np.mean(data, axis=0)
data_scaled = data - data_mean

# Tính ma trận hiệp phương sai
covariance_matrix = compute_covariance_matrix(data_scaled)

# Tính các vectơ riêng và giá trị riêng
eigenvalues, eigenvectors = find_eigenvalues_and_eigenvectors(covariance_matrix)

# Sắp xếp các giá trị riêng theo thứ tự giảm dần
sorted_eigenvalues, sorted_eigenvectors = sort_eigenvalues_and_eigenvectors(eigenvalues, eigenvectors)

# Chọn số lượng thành phần chính để giữ lại ít nhất 80% thông tin
total_variance_retained = 0.0
num_components = 0
for i in range(len(sorted_eigenvalues)):
    total_variance_retained += sorted_eigenvalues[i]
    num_components += 1
    if total_variance_retained / np.sum(sorted_eigenvalues) >= 0.8:
        break

# Chiếu dữ liệu vào các thành phần chính đã chọn
projected_data = project_data(data_scaled, sorted_eigenvectors, num_components)
```

```python
print("Số lượng thành phần chính được giữ lại:", num_components)
print("Tổng phần trăm phương sai giữ lại:", total_variance_retained / np.sum(sorted_eigenvalues))
```

### Thực hiện 2 phương pháp Naive Bayes, hồi quy Logistic trên giữ liệu giảm chiều

```python
# Đánh giá mô hình Naive Bayes trên dữ liệu đã giảm chiều
nb_pred_train_pca = nb_model_pca.predict(X_train_pca)
nb_pred_test_pca = nb_model_pca.predict(X_test_pca)
nb_accuracy_train_pca = accuracy_score(y_train, nb_pred_train_pca)
nb_accuracy_test_pca = accuracy_score(y_test, nb_pred_test_pca)
print("\nNaive Bayes trên dữ liệu đã giảm chiều:")
print("Accuracy trên tập huấn luyện:", nb_accuracy_train_pca)
print("Accuracy trên tập kiểm tra:", nb_accuracy_test_pca)
```

```python
# Huấn luyện mô hình Logistic Regression trên dữ liệu đã giảm chiều
lr_model_pca = LogisticRegression()
lr_model_pca.fit(X_train_pca, y_train)
```

```python
# Đánh giá mô hình Logistic Regression trên dữ liệu đã giảm chiều
lr_pred_train_pca = lr_model_pca.predict(X_train_pca)
lr_pred_test_pca = lr_model_pca.predict(X_test_pca)
lr_accuracy_train_pca = accuracy_score(y_train, lr_pred_train_pca)
lr_accuracy_test_pca = accuracy_score(y_test, lr_pred_test_pca)
print("\nLogistic Regression trên dữ liệu đã giảm chiều:")
print("Accuracy trên tập huấn luyện:", lr_accuracy_train_pca)
print("Accuracy trên tập kiểm tra:", lr_accuracy_test_pca)
```

```python
# Tính sự thay đổi trong độ chính xác của mô hình Naive Bayes
delta_nb = np.abs(nb_accuracy_test - nb_accuracy_test_pca)

# Tính sự thay đổi trong độ chính xác của mô hình Logistic Regression
delta_lr = np.abs(lr_accuracy_test - lr_accuracy_test_pca)

print("Sự thay đổi trong độ chính xác của mô hình Naive Bayes:", delta_nb)
print("Sự thay đổi trong độ chính xác của mô hình Logistic Regression:", delta_lr)

if delta_nb > delta_lr:
    print("Mô hình Naive Bayes có độ chính xác thay đổi nhiều hơn khi áp dụng dữ liệu đã giảm chiều.")
elif delta_lr > delta_nb:
    print("Mô hình Logistic Regression có độ chính xác thay đổi nhiều hơn khi áp dụng dữ liệu đã giảm chiều.")
else:
    print("Cả hai mô hình có độ chính xác thay đổi tương đương khi áp dụng dữ liệu đã giảm chiều.")
```

## 1. Đọc dữ liệu ảnh, lấy tập dữ liệu 5000 ảnh bất kỳ; giảm số chiều dữ liệu xuống còn 100 chiều (từ 28x28 = 784 chiều ban đầu).

```python
data_path = 'E:\ML\multinomial_logistic_regression'

# path
images_path = os.path.join(data_path, 'train-images-idx3-ubyte.gz')
labels_path = os.path.join(data_path, 'train-labels-idx1-ubyte.gz')

def get_mnist_data(images_path, labels_path, num_images, shuffle=False, _is=True, image_size=28):
    """
    This shuffle param is active when .gz is downloaded at:
    - 'http://yann.lecun.com/exdb/mnist/'
    - This function return random num_images in 60000 or 10000
    """
    # read data
    import gzip # to decompress gz (zip) file
    
    # open file training to read training data
    f_images = gzip.open(images_path,'r')
    
    # skip 16 first bytes because these are not data, only header infor
    f_images.read(16)
    
    # general: read num_images data samples if this parameter is set;
    # if not, read all (60000 training or 10000 test)
    real_num = num_images if not shuffle else (60000 if _is else 10000)
    
    # read all data to buf_images (28x28xreal_num)
    buf_images = f_images.read(image_size * image_size * real_num)
    
    # images
    images = np.frombuffer(buf_images, dtype=np.uint8).astype(np.float32)
    images = images.reshape(real_num, image_size, image_size,)
    
    # Read labels
    f_labels = gzip.open(labels_path,'r')
    f_labels.read(8)
    labels = np.zeros((real_num)).astype(np.int64)
    
    # rearrange to correspond the images and labels
    for i in range(0, real_num):
        buf_labels = f_labels.read(1)
        labels[i] = np.frombuffer(buf_labels, dtype=np.uint8).astype(np.int64)

    # shuffle to get random images data
    if shuffle is True:
        rand_id = np.random.randint(real_num, size=num_images)
        images = images[rand_id, :]
        labels = labels[rand_id,]
        
    # change images data to type of vector 28x28 dimentional
    images = images.reshape(num_images, image_size * image_size)
    return images, labels

images, labels = get_mnist_data(images_path, labels_path, 5000, shuffle=True)

print(images.shape, labels.shape)
```

```python
# Khởi tạo PCA với số lượng thành phần chính là 100
pca = PCA(n_components=100)

# Tiêu chuẩn hóa dữ liệu trước khi áp dụng PCA
scaler = StandardScaler()
images_scaled = scaler.fit_transform(images)

# Áp dụng PCA vào dữ liệu đã tiêu chuẩn hóa
images_pca = pca.fit_transform(images_scaled)

# In ra kích thước của dữ liệu sau khi giảm chiều
print("Kích thước của dữ liệu sau khi giảm chiều:", images_pca.shape)
```
